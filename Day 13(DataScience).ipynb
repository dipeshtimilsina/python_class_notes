```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
normal_dist = np.random.normal(0,1,1000)
```

```python
normal_dist.shape
```

```python
plt.figure(figsize=(10,10))
plt.title("My Plot")
plt.xlabel("Normal Dist")
plt.scatter(x=normal_dist, y=np.zeros(1000))
plt.savefig('plot.png')
plt.show()
```

```python
meu = np.array([20,30])
sigma = np.array([
    [2,5],
    [5,4]
])
```

```python
normal_dist = np.random.multivariate_normal(meu,sigma,1000)
```

```python
normal_dist.shape
```

```python
normal_dist
```

```python
x = normal_dist[:,0]
y = normal_dist[:,1]
```

```python
plt.scatter(x=x,y=y)
plt.show()
```

```python
np.corrcoef(normal_dist.T)
```

```python
np.cov(normal_dist.T)
```

```python
std_scaler = StandardScaler()
std_scaler.fit(normal_dist)
normal_dist_scaled = std_scaler.transform(normal_dist)
```

```python
plt.scatter(x = normal_dist_scaled[:,0],
            y = normal_dist_scaled[:,1])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler, MaxAbsScaler
```

```python
names = ['Normalize(0,1)','RobustScaler','MaxAbs(-1,1)']
```

```python
scalers = [MinMaxScaler(),RobustScaler(),MaxAbsScaler()]
```

```python
scalers
```

```python
for name, scaler in zip(names,scalers):
    print("*************")
    print(name)
    scaler.fit(normal_dist)
    scaled_data = scaler.transform(normal_dist)
    print("Min:", scaled_data.min(axis = 0))
    print("Max:", scaled_data.max(axis = 0))
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
genders_coloumn = np.array(['Male','Female',"Male",'Male','Female'])
```

```python
encoder = LabelEncoder()
```

```python
encoder.fit(genders_coloumn)
encoder.transform(genders_coloumn)
```

```python
one_hot_encoder = OneHotEncoder(sparse=False)
```

```python
one_hot_encoder.fit(genders_coloumn.reshape(-1,1))
one_hot_encoder.transform(genders_coloumn.reshape(-1,1))
```
